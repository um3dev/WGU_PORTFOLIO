## <span style="color:green">B1: Selection of Clustering Technique</span>

The K-Means clustering technique is a widely used unsupervised machine learning algorithm that partitions a set of data points into K clusters. The goal of the K-Means clustering algorithm is to minimize the sum of the squared distances between each data point and its assigned cluster center, also known as the centroid.

The K-Means clustering algorithm works as follows:
<ul class="a">
    <li>The user specifies the number of clusters K and selects the initial centroids for each cluster randomly or using another heuristic.
    <li>Each data point is assigned to the nearest centroid based on its distance from the centroid.
    <li>The centroid of each cluster is updated as the mean of all the data points assigned to that cluster.
    <li>Steps 2 and 3 are repeated until the centroids no longer move or a maximum number of iterations is reached.
</ul> 

For the purpose of determining the optimal number of clusters to use for K-Means, I will employ both the Elbow method and Silhouette analysis in my analysis. To elaborate, the Elbow method involves plotting the within-cluster sum of squares (WCSS) against the number of clusters, and identifying the point where the rate of decrease of WCSS begins to level off, resembling an elbow. On the other hand, the Silhouette analysis entails computing the silhouette score for each data point, which evaluates the similarity of the data point to its own cluster as compared to other clusters. Subsequently, the average silhouette score is calculated for each number of clusters, and the number of clusters with the highest average silhouette score is considered as the optimal number of clusters. This method is preferable as it allows for a more detailed assessment of the quality of clustering in comparison to the Elbow method. Specifically, the silhouette score measures the compactness and separation of the clusters, providing insights into the efficacy of the clustering.

## <span style="color:green">B2: Assumptions of Clustering Technique</span>

The reason for making assumptions is to simplify problems, which can then be solved accurately. When dividing a dataset into clusters, assumptions must be made to define the criteria for a cluster. One of the fundamental assumptions that the K-Means clustering method is based on pertains to the characteristics of the clusters. Specifically, it assumes that the clusters are spherical in shape, which facilitates the differentiation of clusters when the algorithm is implemented. Additionally, the method assumes that the clusters are comparable in size, which enables the establishment of cluster boundaries and the identification of the number of data points in each cluster. The spherical assumption is crucial for forming clusters accurately, but it may lead to unexpected results if violated. However, the assumption about cluster size has its benefits, such as enabling the algorithm to determine the centers of the clusters anywhere, as long as they are initially far apart.

## <span style="color:green">D1: Clustering Technique Analysis</span>

<b>Steps For Clustering Technique Analysis</b>

1. Find the optimal number of clusters using both Elbow Method and Silhouette Method.
2. Perform K-Means Cluster Analysis based on the K value determined in Step 1.
3. Analyze the plot visuals to draw final conclusion.

## Elbow Method

````python
# Using Elbow Method Find Number Of Clusters
# Store SSE Values For Each K In A List

A = df_s[['Tenure','MonthlyCharge']]

with plt.style.context(['dark_background']):

    sse = [] 

    for k in range(1, 10):
        kmeans = KMeans(n_clusters=k, random_state=10)
        kmeans.fit(A)
        sse.append(kmeans.inertia_)
        
    fig, ax = plt.subplots(figsize=(12, 8))
    ax.spines.right.set_visible(False)
    ax.spines.top.set_visible(False)
    ax.axes.grid(False)
    knee = KneeLocator(range(1, 10), sse, curve="convex", direction="decreasing")

    plt.xticks(range(1,10))
    plt.tick_params(axis='both', labelsize=12, pad=15)
    plt.xlabel("Number Of Clusters", family='Arial, sans-serif', fontsize=24, color='white', labelpad=15)
    plt.ylabel("SSE", family='Arial, sans-serif', fontsize=24, color='white', labelpad=15)
    plt.title("Elbow Method", family='Arial, sans-serif', fontweight='bold', fontsize=28, color='white', y=1.05, 
              pad=14, horizontalalignment='center')
    plt.axvline(x=knee.elbow, color='lightgreen', ls='--', lw=4)
    
    # Optimum Point On Knee Plot
    ax.annotate('Optimum Point (K): ({}, {:.0f})'.format(knee.elbow, sse[knee.elbow-1]), xy=(3.2, 1450), 
                xytext=(4, 2500), family='Arial, sans-serif', fontsize=16, 
                    color='white', ha='left', 
                    arrowprops=dict(edgecolor='white', linewidth=5, shrinkA=5, shrinkB=5, 
                                    patchA=None, patchB=None), bbox=dict(pad=10, facecolor='black')) 
    
        
    plt.plot(range(1,10), sse, color='red', marker='o', markersize=15, alpha=0.7, linewidth=2.5)
        
    plt.show()
````

<b>Step 1 - Part A</b>

By plotting the sum of squared errors (SSE) against the number of clusters, the elbow method displays a decreasing trend in SSE as the number of clusters increases. The elbow point, depicted as a green dotted line, is where the SSE values start to decrease at a slower pace. For this dataset, the elbow point occurs at k=3, indicating that this is the optimal number of clusters to use.

## Silhouette Method

````python
# Using Silhouette Method Find Number Of Clusters

with plt.style.context(['dark_background']):

    silhouette_coefficients = []

    for k in range(2, 11):
        kmeans = KMeans(n_clusters=k, random_state=10)
        kmeans.fit(A)
        score = silhouette_score(A, kmeans.labels_)
        silhouette_coefficients.append(score)
    fig, ax = plt.subplots(figsize=(12, 8))
    ax.spines.right.set_visible(False)
    ax.spines.top.set_visible(False)
    ax.axes.grid(False)


    plt.xticks(range(2, 11))
    plt.tick_params(axis='both', labelsize=12, pad=15)
    plt.xlabel("Number Of Clusters", family='Arial, sans-serif', fontsize=24, color='white', labelpad=15)
    plt.ylabel("Silhouette Coefficient", family='Arial, sans-serif', fontsize=24, color='white', labelpad=15)
    plt.title("Silhouette Method", family='Arial, sans-serif', fontweight='bold', fontsize=28, color='white', y=1.05, 
              pad=14, horizontalalignment='center')
    ymax = np.amax(silhouette_coefficients)
    plt.axhline(y=ymax, color='lightgreen', ls='--', lw=4)
        
    ax.annotate('Max Element : {:.3f}' .format(ymax), xy=(2.3, 0.605), xytext=(5.6, 0.56), family='Arial, sans-serif', 
                fontsize=16, 
                    color='white', ha='left', 
                    arrowprops=dict(edgecolor='white', linewidth=5, shrinkA=5, shrinkB=5, 
                                    patchA=None, patchB=None), bbox=dict(pad=10, facecolor='black'))
        
    plt.plot(range(2,11), silhouette_coefficients, color='red', marker='o', markersize=15, alpha=0.7, linewidth=2.5)
````

<b>Step 1. Part B</b>

The silhouette coefficient is maximized at the optimal value, indicated by the highest point on the graph, marked with a green dotted line. Based on this graph, it appears that the maximum silhouette coefficient is slightly above 0.60, which corresponds to using k=2 clusters. This method aims to maximize the distance between the clusters, which is indicated by a silhouette coefficient value closer to 1.00. However, when using k=3 or more clusters, the silhouette coefficient suggests that the clusters may become less distinct and potentially even overlap.

<b>Step 2.</b>

Using standardized data of the features (values between -1 and 1) and the recommended number of clusters from Step 1 a clustering analysis is performed. The elbow method demonstrated k=3 being the optimal number of clusters to use while the Silhouette Method indicated k=2 to be the optimal point.

The silhouette coefficient for k=2 (clusters) is 61% and for k=3 it is 51% with a 10% difference. While the elbow method identified the optimal number of clusters at k=3, there is a possibility of bias since it tends to prioritize creating clusters of equal size over maintaining the unity of the clusters.

## Silhouette Analysis For K-Means Clustering

````python

# Select The Number Of Clusters Based On Silhouette Analysis

A = df_s[['Tenure','MonthlyCharge']]
B = A.to_numpy()
range_n_clusters = [2, 3, 4, 5]

for n_clusters in range_n_clusters:
    # Create 2 Plots Side By Side
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(20, 8)

    # The 1st subplot is the silhouette plot
    # The silhouette coefficient can range from -1, 1 but in this example all
    # lie within [-0.1, 1]
    ax1.set_xlim([-0.1, 1])
    
    # The (n_clusters+1)*10 is for inserting blank space between silhouette
    # plots of individual clusters, to demarcate them clearly.
    ax1.set_ylim([0, len(B) + (n_clusters + 1) * 10])

    # Initialize the clusterer with n_clusters value and a random generator
    # seed of 10 for reproducibility.
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(B)

    # The silhouette_score gives the average value for all the samples.
    # This gives a perspective into the density and separation of the formed clusters
    silhouette_avg = silhouette_score(B, cluster_labels)

    # Silhouette Score For Each Sample
    sample_silhouette_values = silhouette_samples(B, cluster_labels)

    """
    LEFT PLOT: Silhouette Plot For Various Clusters
    """

    y_lower = 10
    for i in range(n_clusters):
        
        # Sort & Aggregate Silhouette Scores For Samples Belonging To Cluster 'i'
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i
        
        cmap = cm.get_cmap('Spectral')
        color = cmap(float(i) / n_clusters)
        
        ax1.fill_betweenx(
            np.arange(y_lower, y_upper),
            0,
            ith_cluster_silhouette_values,
            facecolor=color,
            edgecolor=color,
            alpha=0.7,
        )

        # Label The Silhouette Plots With Cluster Numbers In The Middle
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Generate New 'y_lower' For Next Plot
        y_lower = y_upper + 10  # 10 For The 0 Samples

    ax1.set_title("Silhouette Plot of Clusters", family='Arial, sans-serif', fontweight='bold', fontsize=22, color='yellow', 
              y=1.05, pad=20, horizontalalignment='center')
    ax1.set_xlabel("Silhouette Coefficient Values", family='Arial, sans-serif',  fontweight='bold', fontsize=16, 
                    color='white', labelpad=15)
    ax1.set_ylabel("Cluster Label", family='Arial, sans-serif',  fontweight='bold', fontsize=16, color='white', labelpad=15)

    # Mean Silhouette Score For All Values
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Clear 'yaxis' Labels & Ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])
    
    """
    RIGHT PLOT: Clusters Of Churn Customers
    """
    
    cmap = cm.get_cmap('Spectral')
    colors = cmap(cluster_labels.astype(float) / n_clusters)
    
    ax2.scatter(
        B[:, 0], B[:, 1], marker=".", s=30, lw=0, alpha=0.7, c=colors, edgecolor="k"
    )

    # Label Clusters
    centers = clusterer.cluster_centers_
    
    # Create Pink Circles To Identify Cluster Centroids
    ax2.scatter(
        centers[:, 0],
        centers[:, 1],
        marker="o",
        c="yellow",
        alpha=1,
        s=200,
        edgecolor="k",
    )

    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker="$%d$" % i, alpha=1, s=50, edgecolor="k")

    ax2.set_title("Clusters Of Churn Customers", family='Arial, sans-serif', fontweight='bold', fontsize=22, color='yellow', 
              y=1.05, pad=20, horizontalalignment='center')
    ax2.set_xlabel("Tenure", family='Arial, sans-serif',  fontweight='bold', fontsize=16, color='white', labelpad=15)
    ax2.set_ylabel("Monthly Charge", family='Arial, sans-serif',  fontweight='bold', fontsize=16, color='white', labelpad=15)

    plt.suptitle(
        "Silhouette Analysis For K-Means Clustering \n K = {} / Mean Silhouette Score: {:.3f}" 
        .format(n_clusters,silhouette_avg), family='Arial, sans-serif', 
        fontweight='bold', color='white', y=1.18, horizontalalignment='center', fontsize=28)
    
plt.show()
````

The silhouette score is used to measure how dense and well-separated the clusters are from each other. The value of the Silhouette score will vary from -1 to 1. If the score is 1, the cluster is dense and well separated than other clusters. A value near 0 represents overlappping clusters with data points (customers) very close to the decision boundary of the neighboring clusters. A negative score from -1 to 0 indicated that the data points might have got assigned to the wrong cluster. The plot on the left provides a visual of cluster thickness to demonstrate the size of each cluster with the red dotted line indicating the Mean Silhouette Score. The plot on the right shows colored coded clusters based on values of K to demonstrate if clusters are well separated or overlapping.
